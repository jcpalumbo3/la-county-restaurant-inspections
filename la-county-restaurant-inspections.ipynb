#### Step 2.1: Overview of the Dataset

```python
def print_barplot(df, var, topk=10):
    tbl = pd.DataFrame(df[var].value_counts())
    tbl.columns = ['Count']
    tbl['Percent'] = tbl['Count']/tbl['Count'].sum()
    
    if topk>len(tbl):
        topk = len(tbl)
    tbl[:topk].plot.bar(y='Percent', figsize=(10,5))
    plt.xlabel(var)
    plt.ylabel('Percent')
    plt.show()
```

```python
print(inspections['score'].describe())
```

```python
print_barplot(inspections, 'facility_city', 10)
```

```python
print_barplot(inspections, 'grade')
```

```python
print_barplot(inspections, 'pe_description')
```

```python
print_barplot(violations, 'violation_description')
```
